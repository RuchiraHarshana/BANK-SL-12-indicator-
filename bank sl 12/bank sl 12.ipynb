```python
freefolt=[15.39,41.43,9.28,39.93,45.02,15.93,60.43,24.44,24.64,35.68,41.9,9.04]
```

```python


fl_df = pd.DataFrame()
fl_df['Date'] = full_df['Date']
fl_df['amana_b_x_v_freefolt'] = full_df['amana_b'] * freefolt[0]
fl_df['Camb_b_x_v_freefolt'] = full_df['Camb_b'] * freefolt[1]
fl_df['cargills_b_x_v_freefolt'] = full_df['cargills_b'] * freefolt[2]
fl_df['dfcc_b_x_v_freefolt'] = full_df['dfcc_b'] * freefolt[3]
fl_df['hnb_b_x_v_freefolt'] = full_df['hnb_b'] * freefolt[4]
fl_df['nations_t_b_x_v_freefolt'] = full_df['nations_t_b'] * freefolt[5]
fl_df['ndb_b_x_v_freefolt'] = full_df['ndb_b'] * freefolt[6]
fl_df['pan_ashia_b_x_v_freefolt'] = full_df['pan_ashia_b'] * freefolt[7]
fl_df['snasa_b_x_v_freefolt'] = full_df['snasa_b'] * freefolt[8]
fl_df['seylan_b_x_v_freefolt'] = full_df['seylan_b'] * freefolt[9]
fl_df['smap_b_x_v_freefolt'] = full_df['smap_b'] * freefolt[10]
fl_df['union_b_x_v_freefolt'] = full_df['union_b'] * freefolt[11]

fl_df.head()
```

```python


new_df = pd.DataFrame()

new_df['Date'] = fl_df['Date']
new_df['amana_b_x_v_freefolt'] = fl_df['amana_b_x_v_freefolt'] * df_bank_top12['Market Capitalization'].iloc[0]
new_df['Camb_b_x_v_freefolt'] = fl_df['Camb_b_x_v_freefolt'] * df_bank_top12['Market Capitalization'].iloc[1]
new_df['cargills_b_x_v_freefolt'] = fl_df['cargills_b_x_v_freefolt'] * df_bank_top12['Market Capitalization'].iloc[2]
new_df['dfcc_b_x_v_freefolt'] = fl_df['dfcc_b_x_v_freefolt'] * df_bank_top12['Market Capitalization'].iloc[3]
new_df['hnb_b_x_v_freefolt'] = fl_df['hnb_b_x_v_freefolt'] * df_bank_top12['Market Capitalization'].iloc[4]
new_df['nations_t_b_x_v_freefolt'] = fl_df['nations_t_b_x_v_freefolt'] * df_bank_top12['Market Capitalization'].iloc[5]
new_df['ndb_b_x_v_freefolt'] = fl_df['ndb_b_x_v_freefolt'] * df_bank_top12['Market Capitalization'].iloc[6]
new_df['pan_ashia_b_x_v_freefolt'] = fl_df['pan_ashia_b_x_v_freefolt'] * df_bank_top12['Market Capitalization'].iloc[7]
new_df['snasa_b_x_v_freefolt'] = fl_df['snasa_b_x_v_freefolt'] * df_bank_top12['Market Capitalization'].iloc[8]
new_df['seylan_b_x_v_freefolt'] = fl_df['seylan_b_x_v_freefolt'] * df_bank_top12['Market Capitalization'].iloc[9]
new_df['smap_b_x_v_freefolt'] = fl_df['smap_b_x_v_freefolt'] * df_bank_top12['Market Capitalization'].iloc[10]
new_df['union_b_x_v_freefolt'] = fl_df['union_b_x_v_freefolt'] * df_bank_top12['Market Capitalization'].iloc[11]

new_df.head()
```

```python


market_cap_sum = df_bank_top12['Market Capitalization'].sum()
print(f"The sum of market capitalizations in df_bank_top12 is: {market_cap_sum}")
```

```python


new_df_sum = new_df.iloc[:, 1:].sum(axis=1)
new_df['Total_Market_Cap'] = new_df_sum

bank_sl_12 = pd.DataFrame()
bank_sl_12['Date'] = new_df['Date']
bank_sl_12['Market_Share'] = new_df['Total_Market_Cap'] / market_cap_sum

bank_sl_12.head()
```

```python


import matplotlib.pyplot as plt


bank_sl_12['Date'] = pd.to_datetime(bank_sl_12['Date'])
bank_sl_12.set_index('Date', inplace=True)


plt.figure(figsize=(12, 6))
plt.plot(bank_sl_12['Market_Share'])


plt.title('Bank SL 12')
plt.xlabel('Date')
plt.ylabel('Market Share')


plt.show()
```
